# 2. Calibrate Camera
Get the camera parameters. Need these (fx, fy, cx, cy) to undistort images and more accurately calculate pose

```python
from __future__ import print_function
import numpy as np
import cv2
import argparse
import os
from matplotlib import pyplot as plt



def get_calibration_parameters(images_dir, rows, cols, size, show_detections=True):
    """This function receives several chessboard photos taken at different angles, 
    and calculates the calibration parameters specific to the camara being used. 
    These parameters are then used to undistort images."""
    
    

    path = images_dir

    options_dict = dict()
    options_dict['filenames'] = [os.path.join(path, f) for f in os.listdir(images_dir)]
    options_dict['rows'] = rows
    options_dict['cols'] = cols
    options_dict['size'] = size
    options_dict['show_detections'] = show_detections
    
    patternsize = (options_dict['rows'], options_dict['cols'])
    
    # termination criteria
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)

    sz = options_dict['size']

    x = np.arange(patternsize[0])*sz
    y = np.arange(patternsize[1])*sz

    xgrid, ygrid = np.meshgrid(x, y)
    zgrid = np.zeros_like(xgrid)
    opoints = np.dstack((xgrid, ygrid, zgrid)).reshape((-1, 1, 3)).astype(np.float32)

    imagesize = None

    win = 'Calibrate'
    cv2.namedWindow(win)

    ipoints = []


    for filename in options_dict['filenames']:

        rgb = cv2.imread(filename)
        
        if rgb is None:
            print('warning: error opening {}, skipping'.format(filename))
            continue

        cursize = (rgb.shape[1], rgb.shape[0])
        
        if imagesize is None:
            imagesize = cursize
        else:
            assert imagesize == cursize

        print('loaded ' + filename + ' of size {}x{}'.format(*imagesize))

        if len(rgb.shape) == 3:
            gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
        else:
            gray = rgb


        retval, corners = cv2.findChessboardCorners(gray, patternsize)

        if options_dict['show_detections']:
            display = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)
            cv2.drawChessboardCorners(display, patternsize, corners, retval)
            cv2.imshow(win, display)
            # while cv2.waitKey(5) not in range(128): pass

        if not retval:
            print('warning: no chessboard found in {}, skipping'.format(filename))
        else:
            print('pattern found in {}'.format(filename))
            ipoints.append( corners )
            
            corners2=cv2.cornerSubPix(gray,corners, (11,11), (-1,-1), criteria)
#             imgpoints.append(corners)
            # Draw and display the corners
            cv2.drawChessboardCorners(rgb, (9,6), corners2, retval)
            cv2.imshow('rgb', rgb)
            cv2.waitKey(500)


    flags = (cv2.CALIB_ZERO_TANGENT_DIST |
             cv2.CALIB_FIX_K1 |
             cv2.CALIB_FIX_K2 |
             cv2.CALIB_FIX_K3 |
             cv2.CALIB_FIX_K4 |
             cv2.CALIB_FIX_K5 |
             cv2.CALIB_FIX_K6)

    opoints = [opoints] * len(ipoints)

    retval, K, dcoeffs, rvecs, tvecs = cv2.calibrateCamera(opoints, ipoints, imagesize,
                                                           None, None)

#     assert( np.all(dcoeffs == 0) )
    
    fx = K[0,0]
    fy = K[1,1]
    cx = K[0,2]
    cy = K[1,2]

    params = (fx, fy, cx, cy)

    print()
    print('all units below measured in pixels:')
    print('  fx = {}'.format(K[0,0]))
    print('  fy = {}'.format(K[1,1]))
    print('  cx = {}'.format(K[0,2]))
    print('  cy = {}'.format(K[1,2]))
    print()
    print('pastable into Python:')
    print('  fx, fy, cx, cy = {}'.format(repr(params)))
    print()
    
    return(retval, K, dcoeffs, rvecs, tvecs)




```

```python
calibration_parameters = get_calibration_parameters('/home/sarita/Documents/SeniorD/apriltag_detection/calibration_images/webcam/chessboard/large_hor',9, 6, 1.0)
ret, mtx, dist, rvecs, tvecs = calibration_parameters
```

### 2.1 (Optional) Undistort apriltag images:

This method reads an image from image_path, and saves an undistorted copy in 'undistorted' folder within that directory.

**For viewing and debbuging purposes only. Method 2 for pose estimation includes image "undistortion" within it, so no need to feed in an "undistorted" image.**

```python
image_path = '/home/sarita/Documents/SeniorD/calibration_images/webcam/apriltags/original'

for filename in os.listdir(image_path):
    print("Currently reading", filename)
    img = cv2.imread(os.path.join(image_path, filename))
    h,  w = img.shape[:2]
    newcameramtx, roi=cv2.getOptimalNewCameraMatrix(mtx, dist, (w,h), 1, (w,h))

    # undistort
    dst = cv2.undistort(src=img, cameraMatrix=mtx, distCoeffs=dist, newCameraMatrix=newcameramtx)
    # crop the image
    x, y, w, h = roi
    dst = dst[y:y+h, x:x+w]
    cv2.imwrite(str(image_path+ "/undistorted/"+"".join(filename.split('.')[:-1]) + '_undistorted.jpg'), dst)
```

### 3.1 Method 1: Requires additional calculation of pose based on apriltag corners detection
Using the apriltag module on Pypi: https://pypi.org/project/apriltag/
Provides no out of the box pose estimation methods. 
I would have to write the similar triangle math to estimate pose. Better look for an off the shelf solution.

**check out method 2, is much better**

```python
import apriltag
import cv2

images_path = '/home/sarita/Documents/SeniorD/calibration_images/webcam/apriltags/undistorted'
filename = 'april7_undistorted.jpg'

img = cv2.imread(os.path.join(images_path, filename), cv2.IMREAD_GRAYSCALE)
detector = apriltag.Detector()
result = detector.detect(img)

print("The ID of ", filename, "is", result[0].tag_id, "\n\n")
print("Corners Found are: \n", result[0].corners)


tl_hor = int(round(result[0].corners[0][0]))
tl_ver = int(round(result[0].corners[0][1]))

tr_hor = int(round(result[0].corners[1][0]))
tr_ver = int(round(result[0].corners[1][1]))

br_hor = int(round(result[0].corners[2][0]))
br_ver = int(round(result[0].corners[2][1]))

bl_hor = int(round(result[0].corners[3][0]))
bl_ver = int(round(result[0].corners[3][1]))

#draw corners in image
#must provide vertical,horizontal coordinates
img[tl_ver, tl_hor] = 255
img[tr_ver, tr_hor] = 255
img[br_ver, br_hor] = 255
img[bl_ver, bl_hor] = 255

cv2.imwrite(str(image_path+ "/"+ "".join(filename.split('.')[:-1]) + '_corners.jpg'), img)
print(img.shape)

```

### 3.2 Method 2: Includes pose estimation

On it they reference these python bindings by duckietown:

Cloned and followed readme to build the C++ Apriltags library on terminal.

**A note: looks like there are several hyperparameters to the detector that may improve accuracy when tuned. Must simulate environment of competition and test on different combinations of hyperparameters.**

Outputs pose as:
* A 3x3 rotation Matrix of the pose estimate
* A 3x1 vector translation of the pose estimate
* A scalar indicating the object-space error of the estimate


```python
import sys
path_to_library = '/home/sarita/Documents/SeniorD/apriltag_detection/apriltags3-py' #Paste here the path to apriltags3-py you cloned
sys.path.insert(1, path_to_library)
import apriltags3
import cv2

# Detector parameters explained in https://github.com/duckietown/apriltags3-py
at_detector = apriltags3.Detector(searchpath=[str(path_to_library)+'/apriltags/lib'], 
                       families='tag36h11',
                       nthreads=1,
                       quad_decimate=1.0,
                       quad_sigma=0.0,
                       refine_edges=1,
                       decode_sharpening=0.25,
                       debug=0)

# path to apriltag image to load and detect
image_path = "/home/sarita/Documents/SeniorD/apriltag_detection/calibration_images/webcam/apriltags/original/april7.jpg" 
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

camera_parameters = [680.9565272862395, 678.9601430884998, 320.50156915761715, 233.43506067483017]

tags = at_detector.detect(img, estimate_tag_pose=True, camera_params=camera_parameters, tag_size=0.172)

print(tags)
```
